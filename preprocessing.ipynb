```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
my_csv = pd.read_csv("csv_files\EEG_data.csv")
column = my_csv["Theta"]
print(column)
```

```python
df = pd.DataFrame(my_csv,columns=['Theta'])
old_max = df.max()
print(df.idxmax())
old_min = df.min()
print(df.idxmin() )
```

```python
new_min = 100
new_max = 6000
for old_value,row in df.iterrows():
    new_value = ( (row - old_min) / (old_max - old_min) ) * (new_max - new_min) + new_min
    df.at[old_value, 'Theta'] = new_value

```

```python
df.to_csv("csv_files\Freq_Theta.csv", header=False)
```
