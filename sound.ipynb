```python
import pandas as pd
my_csv = pd.read_csv("csv_files\Freq_Theta.csv")
df = pd.DataFrame(my_csv,columns=['Theta'])
```

```python
from math import log2, pow

A4 = 440
C0 = A4*pow(2, -4.75)
name = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]

def pitch(freq):
 h = round(12*log2(freq/C0))
 octave = h // 12
 n = h % 12
 return name[n] + str(octave)

s=''
for old_value,row in df.iterrows():
    s=s+(' ' + pitch(row))
print(s)
```

```python
f = open("Music_notes.txt", "w")
f.write(s)
f.close()
```

```python
from pyknon.genmidi import Midi
from pyknon.music import NoteSeq

notes1 = NoteSeq(s)
midi = Midi(1, tempo=132)
midi.seq_notes(notes1, track=0)
midi.write("Brain_Waves.mid")
```
